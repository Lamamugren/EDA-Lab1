### Bivariate Visualization 
* Plot two Charts
* Provide each charts with at least two insights 

```python
plt.scatter(df["PositionID"], df["Salary"])
plt.xlabel("Position ID")
plt.ylabel("Salary")
plt.title("Salary vs. Position ID")
plt.show()
```

## Scatter plot of Salary vs. Position ID
The scatter plot "Salary vs. Position ID" provides insights into the relationship between an employee's position within the company and their salary. Here are some of the insights that can be drawn from this plot:

Clustering: There is a clear clustering of points in the plot, which indicates that there is a general relationship between salary and position within the company.

Trend: There appears to be a positive trend within the plot's (clusterings), with higher position IDs generally corresponding to higher salaries. This suggests that employees with higher positions within the company are likely to earn more than those in lower positions.

Outliers: There are a few points in the plot that fall outside of the general clustering and trend. These points could represent employees with salaries that are higher or lower than would be expected based on their position within the company. Further investigation may be needed to understand the reasons for these outliers.

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="Department", y="Salary", data=df)
plt.xlabel("Department")
plt.ylabel("Salary")
plt.title("Salary by Department")
plt.show()
```

## Boxplot of Salary by Department


This boxplot shows the Comparison of salaries across departments. The plot allows you to easily compare the salaries across departments and identify which departments have higher median salaries, which ones have a wider range of salaries and which ones have outliers. The dots above and below the whiskers in the plot represent outliers in salary distribution. These outliers might indicate employees with exceptional performance or employees who have been with the company for a long time.The box in the plot represents the IQR, which gives an idea of the distribution of salaries within a department. The width of the box can give an indication of how dispersed the salaries are within a department. For example, a wider box indicates a wider range of salaries within a department.
